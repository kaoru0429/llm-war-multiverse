# 03_talent_migration.ipynb: AI 人才流動分析

視覺化從 OpenAI 到 Anthropic 的關鍵人才流失，這反映了價值觀（Safety vs Speed）的分歧。

```python
import pandas as pd
import plotly.graph_objects as go

# 讀取數據
df = pd.read_csv('../data/talent_flow.csv')

# 構建 Sankey Diagram 數據
all_nodes = list(pd.concat([df['from_company'], df['to_company']]).unique())
node_map = {node: i for i, node in enumerate(all_nodes)}

sources = df['from_company'].map(node_map)
targets = df['to_company'].map(node_map)
values = [1] * len(df)  # 每個人代表 1 個單位
labels = df['person_name'] + ' (' + df['role'] + ')'

fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = all_nodes,
      color = "blue"
    ),
    link = dict(
      source = sources,
      target = targets,
      value = values,
      label = labels
  ))])

fig.update_layout(title_text="OpenAI 人才流失 (The Exodus)", font_size=10)
fig.show()
```

## 人才流動洞察

圖表清晰顯示了單向的人才流動：從 **OpenAI** 流向 **Anthropic**。特別是 2021 年的創始團隊出走和 2024 年的安全團隊負責人（Jan Leike, John Schulman）出走，對 OpenAI 的安全文化構成了重大打擊。
