# 02_performance_analysis.ipynb: 性能與基準測試對比

分析 Kimi K2, GPT-5, DeepSeek V3 等模型在關鍵基準測試上的表現。

```python
import pandas as pd
import plotly.express as px

# 讀取數據
df = pd.read_csv('../data/benchmarks.csv')

# 繪製分組長條圖
fig = px.bar(df, x='benchmark_name', y='score', color='model_name',
             barmode='group',
             title='主要模型基準測試得分對比',
             labels={'score': '得分 (%)', 'benchmark_name': '基準測試'},
             text='score')

fig.update_traces(textposition='outside')
fig.show()
```

## 分析

1. **推理能力**: Kimi K2 Thinking 在 `Humanity's Last Exam` 和 `HMMT 2025` 上表現優異，甚至超越了 GPT-5。
2. **編碼能力**: GPT-5 在 SWE-bench 上仍保持領先，但 Kimi K2 緊隨其後，差距極小。
